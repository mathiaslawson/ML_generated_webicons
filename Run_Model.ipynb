```python
from diffusers import StableDiffusionPipeline
import torch

model_id = "mathiaslawson/web-icons"
pipe = StableDiffusionPipeline.from_pretrained(model_id, torch_dtype=torch.float16).to("cuda") # Load model on GPU

prompt = "a icon of lion<outline-icon> "
generator = torch.Generator("cuda") # Generator is already on cuda

images = pipe(prompt).images  # Get the generated images directly

if images:
    image = images[0]
    print(image)
    image.save("test_icon.png")
else:
    print("No images were generated.")
```
